```python
import pandas
df = pandas.read_csv( "creditCardDefault.csv" )
df.shape
df
```

```python
df.describe()
hist = df.Payment.hist()
```

```python
import numpy
rng = numpy.random.RandomState( 28 )
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split( df, df.Payment, random_state=rng, test_size=0.5 )
```

```python
lstNum = [ 'Limit_Bal', 'Age', 'Bill_Amt1', 'Pay_Amt1' ]
lstCat = [ 'Card', 'Marriage', 'Pay_0' ]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
pipeNum = Pipeline( [
    ('selector', ColumnTransformer([ ('selector', 'passthrough', lstNum ) ] )),
    ('scaler', StandardScaler() )
])
```

```python
from sklearn.preprocessing import OneHotEncoder
pipeCat = Pipeline([
    ('selector', ColumnTransformer([ ('selector', 'passthrough', lstCat ) ] )),
    ('encoder', OneHotEncoder( dtype=int, drop='first' ) )
])
```

```python
from sklearn.pipeline import FeatureUnion
preprocessor = FeatureUnion([
    ('cat', pipeCat ),
    ('num', pipeNum )
])
```

```python
from sklearn.linear_model import LogisticRegression
pipe = Pipeline([
    ('preprocessor', preprocessor ),
    ('model', LogisticRegression( max_iter = 200 ) )
    ])
```

```python
pipe.fit( X_train, y_train )
```

```python
import sklearn.metrics
predTrain = pipe.predict( X_train )
```

```python
# sklearn.metrics.r2_score( y_train, predTrain )
# wont work, as we cant use numeric oriented measures in a classification (logistic) problem
```

```python
pandas.value_counts( y_train )
pandas.value_counts( predTrain )
```

```python
sklearn.metrics.confusion_matrix( y_train, predTrain )
```

```python
sklearn.metrics.accuracy_score( y_train, predTrain )
sklearn.metrics.cohen_kappa_score( y_train, predTrain )
sklearn.metrics.precision_score( y_train, predTrain, pos_label="Paid" )
sklearn.metrics.recall_score( y_train, predTrain, pos_label="Paid" )
sklearn.metrics.f1_score( y_train, predTrain, pos_label="Paid" )
```

```python
print( sklearn.metrics.classification_report( y_train, predTrain ) )
```
