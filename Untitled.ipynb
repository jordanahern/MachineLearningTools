```python
import pandas
df = pandas.read_csv( "creditCardDefault.csv" )
```

```python
lstNum = [ 'Limit_Bal', 'Bill_Amt1' ]
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
pipeNum = Pipeline( [
    ("selector", ColumnTransformer([ ("selector", "passthrough", lstNum ) ] )),
    ('scaler', StandardScaler() )
    ])
```

```python
from sklearn.cluster import KMeans
pipeSimpleKmeans2 = Pipeline([
    ('preprocessor', pipeNum ),
    ('model', KMeans( n_clusters=2 ) )
    ])
```

```python
from sklearn.cluster import AgglomerativeClustering
from mlxtend.preprocessing import DenseTransformer
pipeAgg8 = Pipeline([
    ('preprocessor', pipeNum ),
    ('todense', DenseTransformer() ),
    ('model', AgglomerativeClustering( n_clusters=8 ) )
    ])
```

```python
pipeAgg8.fit( df )
df[ 'clusterAgg8' ] = pipeAgg8[ 'model' ].labels_
```

```python
plot = df.boxplot( column=[ 'Limit_Bal' ], by =[ 'clusterAgg8' ] )
```

```python
plot = df.boxplot( column=[ 'Age' ], by = [ 'clusterAgg8' ] )
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
plot = ConfusionMatrixDisplay.from_predictions( df.clusterKmeans8, df.clusterAgg8 )
```

```python
pandas.crosstab( df.clusterAgg8, df.Payment, margins=True, normalize ="index" ).round( 2 )
```
